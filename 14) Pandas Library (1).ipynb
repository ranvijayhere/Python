## ➩ __ffill and bfill__ ( forwardfill and backwardfill )
#### _ffill ( forward fill ) and bfill ( backward fill ) are methods used to fill missing values ( NaN or NA ) in a DataFrame or Series by propagating existing valid observations._
#### _If you have a sequence [ 1 , NaN , NaN , 4 ], ffill() would result in [ 1 , 1 , 1 , 4 ]._
####  _If you have a sequence [ 1 , NaN , NaN , 4 ], bfill() would result in [ 1 , 4 , 4 , 4 ]._

```python
                                                     # fill and bfill 
health['Age'] = health['Age'].ffill()
health.head()
```

```python
# lets make another column to categorize books based on their user ratings
# 1. Good    # 2. Must Read     # 3 Average

def category(r):
    if r<4:
        return "Average"
    elif r>=4 and r<=4.5:
        return 'Good'
    else:
        return 'Must Read'

                         # lets apply this function to userrating column
df['Category'] = df['UserRating'].apply(category)
df.head(45)
```

# ➩ groupby () method
### ___groupby()__ function is a powerful tool used to split a DataFrame into groups based on one or more columns, allowing for efficient data analysis and aggregation. It follows a __"split-apply-combine" strategy, where data is divided into groups,__ a function is applied to each group, and the results are combined into a new DataFrame_
### _After splitting , a function is applied to each individual group. This function can be an aggregation   (e.g., __sum() , mean() , count() , min() , max() )__ , a transformation (e.g., transform()), or a custom function defined by the user (using apply())._

```python
gb = df.groupby('Year')

gb.first()                    # first entry of each category
```

```python
gb = df.groupby('Year')['Genre']
gb.first()
```

```python
gb = df.groupby('Year')[['Name', 'UserRating']]
gb.first()
```

 ### _Which Year has the average user rating ⮯_

```python
gb = df.groupby('Year')['UserRating'].mean()
gb
```

### _Which Year has the highest user rating_ ⮯

```python
import numpy as np

gb = np.max ( df.groupby('Year')['UserRating'].mean().values )
gb
```
