# __➩ Working with Excel file__

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('C:/Users/woake/Downloads/2020-2025-education-in-danger-incident-data.xlsx')
df.head()
```

```python
df.shape
```

```python
                               # checking the missing values
df.isnull().sum()
```

```python
                            # dropping the unneccesary columns

df = df.drop(['Event Description','SiND Event ID','Known Student Kidnap Or Arrest Outcome',
              'Latitude', "Longitude", 'Known Educators Kidnap Or Arrest Outcome', 'Geo Precision', 'Country ISO' ], axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.reset_index()
```

```python
                 # how many teachers got killed in total

df['Educators Killed'].sum()
```

```python
                       # top 5 highest number of students got killed by country

gb = df.groupby('Country')['Students Killed'].sum().sort_values(ascending=False)
gb.head(5)
```

```python
                        # highest number of Educators got killed by country

gb = df.groupby('Country')['Educators Killed'].sum().sort_values(ascending=False)
gb
```

```python
                     # top 5 highest number of students got killed by country

gb = df.groupby('Country')['Students Killed'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,3))
gb.head(5).plot(kind='bar')
plt.show()
```
